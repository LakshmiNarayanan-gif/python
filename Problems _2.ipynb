```python
## program to check if the list is sorted or not
```

```python
a=[1,3,4,7,9,4,2,3,3,44,2,244,6,5]
```

```python
a.sort()
```

```python
upper_limit=a[0]
lower_limit=a[0]
c=[]
for i in a:
    if i > upper_limit:
        c+=[i]
    else:
        c+=[i]
print(c)
if a==c:
    print( 'list is sorted')
else:
    print ( 'list is unsorted')
        
```
