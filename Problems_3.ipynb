### A dictionary with Student names as keys and their marks as values.

Students who scored more than 50 is printed



```python
def Marks(**Students):
    ''' The Name of the student and his marks were given.It prints the name of the student who scored more than 50'''
    
    for i in Students:
        if Students[i]> 50:
            print (i)
```

### Anagram
A word or Phrase that can be rearranged to form a new word or phrase. The length of the anagrams is not always same

```python
def Anagrams(a,b):
    ''' To find out whether the given words or phrases are anagrams or not'''
    c=a.replace(' ','').lower()
    d=b.replace(' ','').lower()
    if len(c) != len(d):
        print ('It is not anagram')
    else :
        e=dict.fromkeys(c,0)
        for i in c:
            if i in e:
                e[i]+=1
        f=dict.fromkeys(d,0)
        for i in d:
            if i in f :
                f[i]+=1
        
        if e==f:
            print(a,' and ',b , ' are anagrams')
        else :
            print(a,' and ',b , ' are not anagrams')
```

```python
Anagrams('live','Evil')
```

```python
Anagrams('A gentleman','Elegantman')
```

```python
def Merged_dict(a,b):
    """ Give two dictionary to be merged into one"""
    for i in range(len(b)):
        c=b.popitem()
        a.update([c])
    print(a)
    
```
